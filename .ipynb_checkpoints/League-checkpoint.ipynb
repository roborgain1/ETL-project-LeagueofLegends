```python
import os
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

#first LoL db, using platinum+ rankings
url = 'https://champion.gg/statistics/overview?queue=ranked-solo-duo&rank=platinum_plus&region=world'
response = requests.get(url)
soup = bs(response.text, 'html.parser')
print(soup.prettify())
```

```python
#extract column names
results = soup.find('div', class_="StickyTableHeader-sc-111omlw-0 ldshpo")
print(len(results))
print('-'*20)

league_columns = []
for result in results:
    league_columns.append(result.text)
print(league_columns)
```

```python
#create DataFrame column template
df = pd.DataFrame(columns = list(league_columns))
df = df.rename(columns={"helpTierarrow-dropdown": "Tier"})
df = df.drop(columns=['Counters'])
df = df.drop(columns=['Tier'])
df = df.drop(columns=['Rank'])
df = df.drop(columns=['Matches'])
df
```

```python
#extract all rows to iterate through
results = soup.find_all('div', class_="champion-row")
print(f'Number of Rows: {len(results)}')

#append rows to DataFrame
for result in results:
    try:
        role = result.find('div', class_="champion-role").text
        role = role.replace("role-", "")
        role = role.capitalize()
        champion = result.find('span', class_="champion-name").text
        win_rate = result.find('div', class_="champion-win-rate").text
        ban_rate = result.find('div', class_="champion-ban-rate").text
        pick_rate = result.find('div', class_="champion-pick-rate").text
        df = df.append({'Champion': champion, 'Role': role, 'Pick Rate': pick_rate, 'Win Rate': win_rate, 'Ban Rate': ban_rate}, ignore_index=True)
    except AtributeError as e:
        print(e)
        
df = df[['Champion', 'Role', 'Pick Rate', 'Win Rate', 'Ban Rate']]
df
```

```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import re

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

#second LoL db, using iron+ rankings
url = 'https://www.leagueofgraphs.com/champions/builds/iron/sr-ranked'
browser.visit(url)
```

```python
html = browser.html
soup = bs(html, 'html.parser')
```

```python
#create seperate DataFrame-column templates
champion_df = pd.DataFrame(columns = {"Champion"})

role_df = pd.DataFrame(columns = {"Role"})

pick_rate_df = pd.DataFrame(columns = {"Pick Rate"})

win_rate_df = pd.DataFrame(columns = {"Win Rate"})

ban_rate_df = pd.DataFrame(columns = {"Ban Rate"})
```

```python
#create/join champion & role dataframes
champions = soup.find_all('span', class_='name')        
for champion in champions:
    champion = champion.text.strip()
    champion_df = champion_df.append({'Champion': champion}, ignore_index=True)
    
roles = soup.find_all('i')
role_df = pd.DataFrame(columns = {'Role'})
for role in roles:
    if not str(role.text) =="":
        role = role.text.strip()
        role_df = role_df.append({'Role': role}, ignore_index=True)
        
first_join_df = champion_df.join(role_df)
```

```python
#create pick rate dataframe & join to previous
pick_rates = soup.find_all('div', class_="progressBarTxt")

for index, pick_rate in enumerate(pick_rates):
    if index % 3 == 0:
        pick_rate = pick_rate.text.strip()
        pick_rate_df = pick_rate_df.append({'Pick Rate': pick_rate}, ignore_index=True)

second_join_df = first_join_df.join(pick_rate_df)
```

```python
#create win rate dataframe & join to previous
win_rates = soup.find_all('div', class_="progressBarTxt")

for index, win_rate in enumerate(win_rates):
    if (index -1) % 3 == 0:
        win_rate = win_rate.text.strip()
        win_rate_df = win_rate_df.append({'Win Rate': win_rate}, ignore_index=True)

third_join_df = second_join_df.join(win_rate_df)
```

```python
#create ban rate dataframe & join to previous
ban_rates = soup.find_all('div', class_="progressBarTxt")

for index, ban_rate in enumerate(ban_rates):
    if (index -2) % 3 == 0:
        ban_rate = ban_rate.text.strip()
        ban_rate_df = ban_rate_df.append({'Ban Rate': ban_rate}, ignore_index=True)

fourth_join_df = third_join_df.join(ban_rate_df)
fourth_join_df
```
